**Loading NLLB-200 Model for Translation**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load the NLLB-200 model and tokenizer
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Define a mapping from human-readable language names to NLLB codes
lang_map = {
    "English": "eng_Latn",
    "Hindi": "hin_Deva",
    "Dogri": "doi_Deva", # Assuming Dogri Devanagari script
    "Punjabi": "pan_Guru", # Assuming Punjabi Gurmukhi script
    "Japanese": "jpn_Jpan"
}

print("NLLB-200 model and tokenizer loaded successfully.")
print(f"Model moved to: {device}")
```

**Translation Function Using NLLB Model**

```python
def translate_text(text: str, source_lang: str, target_lang: str):
    """
    Translates text using NLLB-200 by manually handling tokens and generation.
    """
    try:
        if source_lang not in lang_map or target_lang not in lang_map:
            return f"Error: Unsupported language. Supported: {', '.join(lang_map.keys())}"

        src_code = lang_map[source_lang]
        tgt_code = lang_map[target_lang]

        # Set the source language for the tokenizer
        tokenizer.src_lang = src_code

        # Tokenize the input text
        inputs = tokenizer(text, return_tensors="pt").to(model.device)

        # Correct way to get the target language token ID for NLLB models
        forced_bos_token_id = tokenizer.convert_tokens_to_ids(tgt_code)

        # Generate translation tokens
        generated_tokens = model.generate(
            **inputs,
            forced_bos_token_id=forced_bos_token_id,
            max_length=400
        )

        # Decode the tokens back to string
        translated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
        return translated_text

    except Exception as e:
        return f"Manual Translation Error: {str(e)}"
```

**Translation Function Using Synthetic Dataset**

```python
import pandas as pd

def translate_text(text, source_lang, target_lang):
    """
    Translates text between supported languages using the synthetic dataset.
    """
    try:
        # Load dataset
        df = pd.read_json('synthetic_translations.json')

        # Check if requested languages are in the dataset columns
        valid_langs = df.columns.tolist()
        if source_lang not in valid_langs or target_lang not in valid_langs:
            return f"Error: Unsupported language. Supported: {', '.join(valid_langs)}"

        # Search for the text in the source language column
        match = df[df[source_lang].str.strip() == text.strip()]

        if not match.empty:
            return match.iloc[0][target_lang]
        else:
            return "Error: Text not found in the synthetic dataset."

    except Exception as e:
        return f"An error occurred: {str(e)}"

# --- Verification Tests ---
test_cases = [
    {"text": "Hello, how are you?", "src": "English", "tgt": "Japanese"},
    {"text": "ਸਤਿ ਸ੍ਰੀ ਅਕਾਲ, ਕੀ ਹਾਲ ਹੈ?", "src": "Punjabi", "tgt": "Hindi"},
    {"text": "तुआढ़ा नां के ऐ?", "src": "Dogri", "tgt": "English"},
    {"text": "Unknown text", "src": "English", "tgt": "Japanese"}, # Test for missing text
    {"text": "Hello", "src": "French", "tgt": "English"} # Test for invalid language
]

print("Translation Function Test Results:")
for case in test_cases:
    result = translate_text(case['text'], case['src'], case['tgt'])
    print(f"[{case['src']} -> {case['tgt']}] Input: {case['text']} | Output: {result}")
```

**Frontend Integration Using FastAPI**

```python
from fastapi import FastAPI, Request, Form
from fastapi.templating import Jinja2Templates
from fastapi.responses import HTMLResponse
import os

# Initialize FastAPI app
app = FastAPI()

# Set up template directory (creating it if it doesn't exist for now)
os.makedirs('templates', exist_ok=True)
templates = Jinja2Templates(directory='templates')

@app.get('/', response_class=HTMLResponse)
async def read_root(request: Request):
    """
    Renders the initial index page.
    """
    return templates.TemplateResponse('index.html', {'request': request})

@app.post('/translate', response_class=HTMLResponse)
async def handle_translation(
    request: Request,
    text: str = Form(...),
    source_lang: str = Form(...),
    target_lang: str = Form(...)
):
    """
    Handles the translation request by calling the translate_text function.
    """
    # Calling the previously defined translate_text function
    translated_text = translate_text(text, source_lang, target_lang)

    return templates.TemplateResponse('index.html', {
        'request': request,
        'original_text': text,
        'source_lang': source_lang,
        'target_lang': target_lang,
        'translated_text': translated_text
    })

print('FastAPI application and endpoints have been initialized.')
```

**Loading NLLB Model with Extended Language Support**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load the NLLB-200 model and tokenizer
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Define a mapping from human-readable language names to NLLB codes
lang_map = {
    "English": "eng_Latn",
    "Hindi": "hin_Deva",
    "Punjabi": "pan_Guru",
    "Japanese": "jpn_Jpan",
    "Telugu": "tel_Telu", # Telugu NLLB code
    "Bengali": "ben_Beng", # Bengali NLLB code
    "Chinese (Simplified)": "zho_Hans", # Chinese (Simplified) NLLB code
    "Russian": "rus_Cyrl", # Russian NLLB code
    "Spanish": "spa_Latn"  # Spanish NLLB code
}

print("NLLB-200 model and tokenizer loaded successfully.")
print(f"Model moved to: {device}")
```

**AI Translation Function with Extended Language Verification**

```python
def translate_text(text: str, source_lang: str, target_lang: str):
    """
    Translates text using NLLB-200 by manually handling tokens and generation.
    """
    try:
        if source_lang not in lang_map or target_lang not in lang_map:
            return f"Error: Unsupported language. Supported: {', '.join(lang_map.keys())}"

        src_code = lang_map[source_lang]
        tgt_code = lang_map[target_lang]

        # Set the source language for the tokenizer
        tokenizer.src_lang = src_code

        # Tokenize the input text
        inputs = tokenizer(text, return_tensors="pt").to(model.device)

        # Correct way to get the target language token ID for NLLB models
        forced_bos_token_id = tokenizer.convert_tokens_to_ids(tgt_code)

        # Generate translation tokens
        generated_tokens = model.generate(
            **inputs,
            forced_bos_token_id=forced_bos_token_id,
            max_length=400
        )

        # Decode the tokens back to string
        translated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
        return translated_text

    except Exception as e:
        return f"Manual Translation Error: {str(e)}"

# --- Verification Tests for the AI Translation Function --- (Updated for new languages)
print(f"Testing AI Translation Logic with updated language map:")

# Test English to Japanese
result_jp = translate_text("Hello, how are you?", "English", "Japanese")
print(f"English to Japanese: {result_jp}")

# Test English to Hindi
result_hi = translate_text("What is your name?", "English", "Hindi")
print(f"English to Hindi: {result_hi}")

# Test English to Telugu
result_te = translate_text("Thank you very much.", "English", "Telugu")
print(f"English to Telugu: {result_te}")

# Test English to Chinese (Simplified)
result_zh = translate_text("I love learning new languages.", "English", "Chinese (Simplified)")
print(f"English to Chinese (Simplified): {result_zh}")

# Test Russian to English
result_ru_en = translate_text("Привет, как дела?", "Russian", "English")
print(f"Russian to English: {result_ru_en}")

# Test Spanish to Bengali
result_es_bn = translate_text("Por favor, espera un momento.", "Spanish", "Bengali")
print(f"Spanish to Bengali: {result_es_bn}")
```

**Updating FastAPI with Language Support**

```python
from fastapi import FastAPI, Request, Form
from fastapi.templating import Jinja2Templates
from fastapi.responses import HTMLResponse
import os

# Initialize FastAPI app
app = FastAPI()

# Set up template directory (creating it if it doesn't exist for now)
os.makedirs('templates', exist_ok=True)
templates = Jinja2Templates(directory='templates')

@app.get('/', response_class=HTMLResponse)
async def read_root(request: Request):
    """
    Renders the initial index page.
    """
    # Pass lang_map to the template context
    return templates.TemplateResponse('index.html', {'request': request, 'lang_map': lang_map})

@app.post('/translate', response_class=HTMLResponse)
async def handle_translation(
    request: Request,
    text: str = Form(...),
    source_lang: str = Form(...),
    target_lang: str = Form(...)
):
    """
    Handles the translation request by calling the translate_text function.
    """
    # Calling the corrected AI translation function using NLLB-200
    translated_text = translate_text(text, source_lang, target_lang)

    return templates.TemplateResponse('index.html', {
        'request': request,
        'original_text': text,
        'source_lang': source_lang,
        'target_lang': target_lang,
        'translated_text': translated_text,
        'lang_map': lang_map # Pass lang_map to the template context
    })

print('FastAPI application and endpoints have been initialized, with lang_map passed to templates.')
```

**Restarting and Testing the Web App**

```python
import multiprocessing
import uvicorn
from fastapi.testclient import TestClient
from google.colab.output import eval_js
import time

# 1. Terminate existing background processes running uvicorn
try:
    if 'proc' in globals() and proc.is_alive():
        proc.terminate()
        proc.join()
        print("Existing server process terminated.")
except Exception as e:
    print(f"Note: Error terminating process: {e}")

# 2. Start the FastAPI application in a new background process
def run_server():
    uvicorn.run(app, host='0.0.0.0', port=8000, log_level='error')

proc = multiprocessing.Process(target=run_server, daemon=True)
proc.start()

# Give the server a few seconds to initialize
time.sleep(3)

# 3. Generate and display the public access URL
proxy_url = eval_js('google.colab.kernel.proxyPort(8000)')
print(f"FastAPI server restarted successfully.")
print(f"Access your translation app here: {proxy_url}")

# 4. Programmatic Verification using TestClient
client = TestClient(app)

print("\n--- Running Programmatic Verification ---")

# Test GET request to root for new title
response_get = client.get('/')
assert response_get.status_code == 200
assert "IBM Translator" in response_get.text
print("GET '/': Success (Status 200, 'IBM Translator' title found)")

# Test POST request for translation with new languages
test_payloads = [
    {"text": "Hello, how are you?", "source_lang": "English", "target_lang": "Japanese"},
    {"text": "Thank you very much.", "source_lang": "English", "target_lang": "Telugu"},
    {"text": "I love learning new languages.", "source_lang": "English", "target_lang": "Chinese (Simplified)"},
    {"text": "Привет, как дела?", "source_lang": "Russian", "target_lang": "English"},
    {"text": "Por favor, espera un momento.", "source_lang": "Spanish", "target_lang": "Bengali"}
]

for payload in test_payloads:
    response_post = client.post('/translate', data=payload)
    assert response_post.status_code == 200
    if "Translation Result:" in response_post.text and "Error" not in response_post.text:
        print(f"POST '/translate' ({payload['source_lang']} to {payload['target_lang']}): Success (Status 200, Translation result found)")
    else:
        print(f"POST '/translate' ({payload['source_lang']} to {payload['target_lang']}): Warning/Failure (Status 200, but translation might be missing or errored in HTML)")

print("--- Verification Completed Successfully ---")
```
